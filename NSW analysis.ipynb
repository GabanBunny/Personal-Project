```python
#Data preparation
#Full time
ftfemale = df2[(df2.Gender == "Female") & (df2["PT/FT"] == 'Full-Time')]
ftmale = df2[(df2.Gender == "Male") & (df2["PT/FT"] == 'Full-Time')]

#Part time
ptfemale = df2[(df2.Gender == "Female") & (df2["PT/FT"] == 'Part-Time')]
ptmale = df2[(df2.Gender == "Male") & (df2["PT/FT"] == 'Part-Time')]

#Year fulltime
ftmaleYear = ftmale.groupby('Year').Headcount.sum()
ftfemaleYear = ftfemale.groupby('Year').Headcount.sum()

#Year partime
ptmaleYear = ptmale.groupby('Year').Headcount.sum()
ptfemaleYear = ptfemale.groupby('Year').Headcount.sum()
```

```python
#Encode
enc = OrdinalEncoder()
dfEncode = df2
#Encode Gender
dfEncode = pd.get_dummies(data = dfEncode, columns = ['Gender','PT/FT'])
dfEncode = dfEncode.join(df2['Gender'])
dfEncode['Gender_Female'] = enc.fit_transform(dfEncode[['Gender_Female']])
dfEncode['Gender_Male'] = enc.fit_transform(dfEncode[['Gender_Male']])


#Encode PT/FT
dfEncode= dfEncode.join(df2['PT/FT'])
dfEncode['PT/FT'] = enc.fit_transform(dfEncode[['PT/FT']])
dfEncode['PT/FT_Full-Time'] = enc.fit_transform(dfEncode[['PT/FT_Full-Time']])
dfEncode['PT/FT_Part-Time'] = enc.fit_transform(dfEncode[['PT/FT_Part-Time']])

dfEncode
```
