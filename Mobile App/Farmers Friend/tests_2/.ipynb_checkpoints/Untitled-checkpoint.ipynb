```python
from datetime import date
import matplotlib.pyplot as plt

base_salary = 70000
salary_sacrifice_super = 10000
age = date.today().year - 2004
```

# ðŸ’° Money In

## Base Salary

```python
print("Base salary:", base_salary)
```

## Super Contributions

```python
employer_super_rate = 0.11  # 11% SG rate for FY2023â€“24
employer_super = base_salary * employer_super_rate
print(f"Employer Super Contribution (11%): ${employer_super:.2f}")
```

## Investment Income

```python
# dividends = float(input("Dividends: "))
# interest_income = float(input("Interest: "))
# capital_gains = float(input("Capital Gains: "))
# rental_income = float(input("Rental Income: "))
# crypto_income = float(input("Crypto Gains: "))
# freelance_income = float(input("Freelance/Side Hustle: "))
# gov_payments = float(input("Gov Payments/Refunds: "))
# gifts = float(input("Gifts/Other: "))
```

# ðŸ’¸ Money Out

```python
# This is the taxable income (used for tax calculation)
taxable_income = base_salary - salary_sacrifice_super

# Apply tax brackets to taxable_income
if taxable_income <= 18200:
    tax = 0
elif taxable_income <= 45000:
    tax = (taxable_income - 18200) * 0.19
elif taxable_income <= 120000:
    tax = (taxable_income - 45000) * 0.325 + 5092
elif taxable_income <= 180000:
    tax = (taxable_income - 120000) * 0.37 + 31167
else:
    tax = (taxable_income - 180000) * 0.45 + 51667
income_tax = tax 
print("Income tax: ", income_tax )
medicare_levy = income_tax*0.02
print("Medicare Levy: ", medicare_levy)
super_tax = 0.15 * salary_sacrifice_super
print("Super Tax: ", super_tax )
```

# ðŸ’²NET CASH AFTER PAY

```python
net_salary = base_salary - salary_sacrifice_super - income_tax - medicare_levy
print("NET CASH AFTER PAY SALARY: ", net_salary)
```

## Living Expense


```python
rent_mortgage = 1300
utilities = 70
groceries = 400
transport = 50 
insurance = 0
eating_out = 100
shopping = 90
fitness = 248
total_living_expense_monthly = rent_mortgage + utilities + groceries + transport + insurance + eating_out + shopping + fitness
```

# ðŸ’²NET CASH AFTER Living Expense

```python
net_cash_after_expense = net_salary - total_living_expense_monthly*12
print("NET CASH AFTER Living Expense: ",net_cash_after_expense )
```

## Wealth Allocation


## Savings

```python
high_growth_savings = 100*12
emergency_savings = total_living_expense_monthly * 3
net_cash_after_expense = net_cash_after_expense - high_growth_savings - emergency_savings
```

## Superannuation

```python
personal_super_contribution_aftertax = 0.01*net_cash_after_expense
net_cash_after_expense = net_cash_after_expense - personal_super_contribution_aftertax
```

```python
equity_budget = (100 - age)/100*net_cash_after_expense
bonds_budget = net_cash_after_expense - equity_budget
net_cash_after_expense = net_cash_after_expense - equity_budget - bonds_budget
```

## Equity

```python
ETFs = (100 - age)/100*equity_budget
Stocks = equity_budget - ETFs
# extra_mortgage = 0
# offset_savings = 0
```

## Debt

```python
gov_bonds =  (100 - age)/100*bonds_budget
corp_bonds =  bonds_budget - gov_bonds
```

# Overall

```python
import matplotlib.pyplot as plt

# Your existing data
labels = ['Emergency Savings', 'Corporate Bonds', 'Government Bonds', 'Stocks', 'ETFs', 'High-Growth Savings','Personal Super Contribution After Tax', 'Total Living Expense Yearly']
values = [emergency_savings, corp_bonds, gov_bonds, Stocks, ETFs, high_growth_savings, personal_super_contribution_aftertax, total_living_expense_monthly*12]

# Sort values and labels by descending value
sorted_data = sorted(zip(values, labels), reverse=True)
values_sorted, labels_sorted = zip(*sorted_data)
legend_labels = [f"{label}: ${value:,.0f}" for label, value in zip(labels_sorted, values_sorted)]

# Plot sorted pie chart
plt.figure(figsize=(8, 8))
plt.pie(values_sorted, labels=labels_sorted, autopct='%1.1f%%', startangle=140)
plt.title("Wealth Allocation Breakdown (Sorted)")
plt.axis('equal')
plt.show()
```
